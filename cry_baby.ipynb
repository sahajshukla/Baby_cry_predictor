```python
from keras.utils import to_categorical
list_outputs = to_categorical(list_outputs)
from sklearn.model_selection import train_test_split
mfccs_train, X_test, y1_train, y_test = train_test_split(mfccs_n, list_outputs, test_size = 0.2)
X_train, X_val, y_train, y_val = train_test_split(mfccs_train, y1_train, test_size = 0.15)
y_val = np.array(y_val)
y_train = np.array(y_train)
y_test = np.array(y_test)
X_train = np.array(X_train)
X_test = np.array(X_test)
X_val = np.array(X_val)
X_train = X_train.reshape((310,13*1))
X_test = X_test.reshape((92, 13*1))
X_val = X_val.reshape((55,13*1))
X_train = np.array(X_train)
X_test = np.array(X_test)
X_val = np.array(X_val)
X_train.shape
X_test.shape
X_val.shape
y_train.shape
```

```python
import keras
from keras import models
from keras import layers
from keras.layers import Dropout
from keras.layers import BatchNormalization
network = models.Sequential()
network.add(layers.Dense(64, activation='relu', input_shape=(13*1,)))
network.add(Dropout(0.5))
network.add(BatchNormalization())
network.add(layers.Dense(32, activation='relu'))
network.add(Dropout(0.5))
network.add(BatchNormalization())
network.add(layers.Dense(32, activation='relu'))
network.add(Dropout(0.5))
network.add(BatchNormalization())
network.add(layers.Dense(5, activation='softmax'))
print(network.summary())
network.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
X_val.shape
y_test.shape
history = network.fit(X_train, y_train,epochs=150,validation_data = [X_val, y_val], batch_size=32)
```

```python
test_loss, test_acc = network.evaluate(X_test, y_test)
```

```python
test_acc
```

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'r', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.legend()

plt.show()

plt.plot(epochs, acc, 'g', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.xticks(epochs)
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
